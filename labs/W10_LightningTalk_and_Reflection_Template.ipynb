## Talk outline (fill these sections)
### 1) Title + your main claim (1 sentence)
**Title:** Beyond Age and Sex: Clinical Predictors of Heart Disease in the UCI Cleveland Dataset.

**Main claim (1 sentence, non-causal unless justified):** Among patients referred for cardiac evaluation, lower maximum heart rate (thalach) and greater exercise-induced ST depression (oldpeak) are independently associated with higher odds of heart disease, even after controlling for age and sex.
### 2) Why do we care (bioengineering relevance, 2–3 sentences)
Heart disease remains the leading cause of death globally. Wearable biosensors
and implantable devices increasingly capture real-time cardiac signals -
including heart rate during exertion and ECG morphology - precisely the
variables that drive our findings. Understanding which physiological signals
carry independent predictive value informs which measurements are worth
engineering into next-generation diagnostic devices. This analysis also
illustrates the full data-science pipeline a bioengineering team would need
to validate a new cardiac biomarker.

### 3) Data (what, where from, sample size, target definition)
- Source:UCI Machine Learning Repository — Cleveland Clinic Foundation
  Heart Disease Dataset (Detrano et al., 1989)
- Rows (n): 303 patients; 297 retained after dropping 8 rows with missing
  values in `ca` and `thal` (< 3% loss)
- Target definition (`disease`): disease` = 1 if original `num` > 0 (any degree of angiographic stenosis), else 0. Disease prevalence: ~54%.
- Key caveats: Referral-only population (not general public); single clinic; collected in 1988; 68% male; physical fitness not recorded.

### 4) Methods (what you did)
- Cleaning policy: Replaced "?" with NaN; coerced numeric columns; created binary disease label; added ca_missing / thal_missing flag columns; dropped 8 incomplete rows. Quality gate (5 assertions) confirmed integrity.
- EDA highlights: thalach is strongly left-separated by disease group (disease patients ~19 bpm lower). Asymptomatic chest pain (cp=4) has the highest disease rate — counterintuitive but clinically documented. Oldpeak is right-skewed with most patients near zero; high values cluster in the disease group. Age confounds the sex–disease relationship (probed by stratification).
- Test/model used: - Permutation test + bootstrap CI for thalach vs disease (W7) - Chi-square + Cramér's V for chest pain type vs disease (W7) - Logistic regression (statsmodels): baseline (age, sex) vs extended (age, sex, thalach, oldpeak, exang) model with odds ratios and 95% CIs (W8)

### 5) Results (one plot + one statistic)
- Figure: Odds ratio forest plot from the extended logistic regression model (age, sex, thalach, oldpeak, exang). Points left of OR=1 are protective; right are risk-increasing. thalach CI is entirely below 1; oldpeak CI is entirely above 1. Both are statistically significant after controlling for the other variables.
- Effect size: thalach: Cohen's d ≈ 1.0 (large); OR ≈ 0.97 per bpm in   
- logistic model -oldpeak: OR ≈ 1.6–2.0 per mm ST depression
- CI: Bootstrap 95% CI for mean thalach difference (no disease - disease): approximately (14, 24) bpm - does not include zero.
- p-value: Permutation p < 0.001 for thalach; chi-square p < 0.001 for cp. Both effects are highly unlikely under the null hypothesis.

### 6) Limitations + what would change your mind
This is a cross-sectional, observational dataset from a single referral clinic in 1988 - it cannot establish causation, and findings may not generalize to healthy populations or contemporary patients. The most important unmeasured confounder is physical fitness: fit patients have higher max heart rates AND lower disease risk independently, which could inflate thalach's apparent protective effect. The dataset is also majority male (~68%), so the sex-specific conclusions are statistically underpowered for women.

What would change my mind: (1) if a prospective cohort study controlling for
fitness level showed thalach loses significance, (2) if the same model fit to
a multi-centre dataset produced substantially different odds ratios, or (3) if
a female-only subgroup analysis showed the thalach effect disappears in women.


### 7) Next step (what data or experiment would you do next?)
Collect a prospective dataset including VO₂max or step-count as a fitness proxy, refit the logistic model, and test whether thalach remains significant after fitness adjustment. Additionally, apply the pipeline to a more diverse dataset to test generalizability across populations and clinical settings.

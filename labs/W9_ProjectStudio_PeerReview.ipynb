### Reviewer summary

1) **Top 2 strengths:**
   - The cleaning pipeline (W3) is fully reproducible and explicitly documented
     with a CHANGELOG - any reviewer can trace every transformation applied to the data.
   - Hypothesis testing (W7) goes beyond p-values: effect sizes (Cohen's d, Cramér's V)
     and bootstrap confidence intervals are reported, which is better practice than
     most introductory analyses.

2) **Top 2 improvements (highest impact):**
   - The logistic regression (W8) does not report model fit metrics (e.g., pseudo-R²,
     AUC, or a confusion matrix). Adding these would clarify how well the model
     actually predicts disease, not just which variables are significant.
   - The stratified analysis (W4, W6) identified age as a potential confounder of the
     sex-disease relationship but never formally tested whether the sex effect disappears
     within age strata. A table of disease rates by sex × age bin would close this gap.

3) **One place the conclusion overreaches:**
   - W8 states that oldpeak is "one of the strongest predictors." This is true within
     this model, but the feature set is limited (5 variables). Without testing all
     available predictors, claiming strongest is an overreach - it should read
     "one of the strongest among the variables tested."
